## 2.  Scrape a single page of non-tabular data



Scrape the following info from <a href="https://www.addictioncenter.com/treatment/top-10-drug-treatment-centers/">this page</a>:
    
- Name of rehab center
- Location
- Phone number
- URL

Export the data into a csv file called ```rehabs.csv```.


````python
ubl = "https://www.addictioncenter.com/treatment/10-top-drug-treatment-centers/"
````

````python
respond = requests.get(ubl)
ic(respond.status_code)
````

````python
sou_p = BeautifulSoup(respond.text, "html.parser")
sou_p
````

````python
print(sou_p.prettify())
````

````python
disorganized = sou_p.find("div", class_= "RehabList")
disorganized
````

````python
type(disorganized)
````

````python
name = disorganized.find_all ("div", class_="RehabListItem__col RehabListItem__col--left")
name
````

````python
name_label = [item.find("h3") for item in name]   ###Sandeep, I do not know why some centers + info are listed twice. 
                                                  ###I checked the page source and it's just like that even though,
                                                ###that's not how it appears on the front end of the website. 
name_label
````

````python
label_txt = [item.get_text() for item in name_label]
label_txt
````

````python
location_label = [item.find("p") for item in name]
location_label
````

````python
location_txt = [item.get_text().replace("Locations in ", "") for item in location_label]
location_txt
````

````python
content = disorganized.find_all ("div", class_="RehabListItem__col RehabListItem__col--right")
content
````

````python
num_label = [item.find("a", class_= "btn btn-primary btn-small") for item in content]
num_label
````

````python
num_txt = [item.get_text() for item in num_label]
num_txt
````

````python
web_label = [item.find("a", class_="btn btn-outline btn-small") for item in content]
web_label
````

````python
web_urls = [item.get("href") for item in web_label]
web_urls
````

````python
len(web_urls)
````

````python
df_cent= pd.DataFrame(list(zip(label_txt, location_txt, num_txt, web_urls)), 
               columns =['Center', "Location",'Phone Number', 'Web Link'])

df_cent
````

````python
filename = "rehabs.csv"
````

````python
df_cent.to_csv(filename, encoding='utf-8', index=False)
````
